```python
import numpy as np
from scipy.interpolate import interp1d
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import time
import random
import sys
from tqdm import tqdm
import matplotlib.pyplot as plt
from munch import Munch
import h5py 
import os

from NL0.initialdata import * 
from NL0.initialdata import idata_params
from io_data import *

params = Munch(dict())
# cosmology
params.omega = 1./3
params.t_ini = 1.
params.H_ini = 2./(3.*(1.+params.omega))/params.t_ini # alpha/t_ini
params.rho_bkg_ini =  3./(8.*np.pi) *params.H_ini**2
params.a_ini = 1
# grid
params.N_r = 0
params.r_max = 0
params.r_is_logarithmic = False
params.sigma_factor = 1
params.dt_multiplier = 0
params.dx = 0
params.dt0 = 0


# sel_style='classic'
# sel_style=''
# sel_style='seaborn-v0_8-bright'
# sel_style='grayscale'

# plt.style.use('classic')
plt.style.use('seaborn-v0_8-dark-palette')
plt.style.use('seaborn-v0_8-talk')
# plt.style.use("seaborn-v0_8-white")
```

```python
mu_list = [
     0.7, 0.8, 0.9,
    1.0,
          ]
t0_list = [1, 2, # 3, 4, 5,
          ]

all_data = []
fig2, axs = plt.subplots(1,2, figsize=(12,5))

for i_mu, mu in enumerate(mu_list): 
    for i_t0, t0 in enumerate(t0_list):
        
        
        plotdir = './NL0/{mu}_{t0}/'.format(mu=int(mu*10**6), t0=int(t0*100))
        h5_filename = './NL0/out_{mu}_{t0}.hdf5'.format(mu=int(mu*10**6), t0=int(t0*100))
        
        try: 
            t_sol, solution, params, idata_params = load_simdata(h5_filename, params, idata_params)
            rm = get_rm(params, idata_params, print_out=0)
            r, initial_state = get_initial_state(params, idata_params)
        except Exception as e:
            print(f"run with mu = {mu} didn't work")
            print(e)
            continue
        
        r_over_rm = r/rm        
        tH = t_sol/(rm**2/4) 
        ti_at_tH = 1 * (rm**2/4)
        rho_bkg_tH = get_rho_bkg(ti_at_tH, params.rho_bkg_ini)
                
            
        mask = (r_over_rm > 0.20) & (r_over_rm < 1.05) 

        # tmax = 8000
        # if mu == 0.4 : rho[t_i>8100] = np.nan
        # # if nu == 0.6825: tmax = 9000
        
        pbh_mass = []

        for i, t_i in enumerate(t_sol):

            # if t_i > tmax: 
            #     drho_max[i]=np.nan
            #     C_max[i] = np.nan
            #     rho_max[i] = np.nan            
            #     continue

            vars_vec = solution[i,:]
            U, R, M, rho = unpack_state(vars_vec, params.N_r)
            U = U[mask]
            R = R[mask]
            M = M[mask]
            rho = rho[mask]
            
            
            rho_bkg = get_rho_bkg(t_i, params.rho_bkg_ini)
            C = compact_function(M, R, rho_bkg)
            drho = rho  /rho_bkg #-1
            rho_m = rho  #/rho_bkg #-1
            
            r_reg = np.linspace(0.3,1.0,1000000)
            C_int = interp1d(r_over_rm[mask], C, kind="linear")
            M_int = interp1d(r_over_rm[mask], M, kind="linear")
            R_int = interp1d(r_over_rm[mask], R, kind="linear")
            
            idx = np.where(C_int(r_reg)>=0.99999)[0]

            if len(idx) > 0: 
                r_ah1 = r_reg[idx[0]]
                r_ah2 = r_reg[idx[-1]]
                
                #print(t_i, np.max(C), M_int(r_ah1) , M_int(r_ah2) )
                
                out = [t_i, M_int(r_ah2), R_int(r_ah2)] 
                pbh_mass.append(out)
        
        pbh_mass = np.array(pbh_mass)
        if pbh_mass.size > 0 : 
            all_data.append([pbh_mass, mu, t0])
        else:
            continue
        # Plot Masses
        
        
        ax = axs[0]
        ax.plot(pbh_mass[:,0]/ pbh_mass[:,0][0]-1   , pbh_mass[:,1]/ pbh_mass[:,1][0], label=f"{t0}", alpha=0.6)
        ax.set_xlabel(r'$ t/t_{\rm form} -1 $')
        ax.set_ylabel(r'$ M/M_{\rm form} $')
        ax.legend()
        # ax.set_yscale('log')
        
        ax = axs[1]
        ax.plot(  (pbh_mass[:,0] - pbh_mass[:,0][0])/ pbh_mass[:,0][0] , pbh_mass[:,1])
        ax.set_xlabel(r'$ t /t_{\rm form} -1$')
        ax.set_ylabel(r'$ M/M_{\rm form} $')
        ax.set_yscale('log')
        
plt.tight_layout()
plt.show()
        
        
        
        

        
        

        
```

In the previous plot we see that the variation of the time for Hubble crossing only changes a proportional factor on masses and time evolutions. Indeed gravitational collapse is conformally invariant. 

Also, the amplitude of the perturbation does affect the masses and mass evolution.  Larger amplitudes results on higher PBH masses, and also the mass evolution growth decays slower.  Still all tend to a constant value. The BH growth should not be computed as ratio to the  mass at formation, as this is depends a lot on the temporal precistion on when the BH is detected. At formation time, the growth of the BH is very large so small deviation, result in large deviation on PBH mass at formation.
